Imbalanced Dataset:

A dataset with an unequal class distribution

```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('creditcard.csv')
```

```python
#printing first 5 rows
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset.shape
```

```python
#distribution of the 2 classes
dataset['Class'].value_counts()
```

This is highly imbalanced datset

0--> normal transaction
1--> fraudlent transaction

```python
legit = dataset[dataset.Class == 0]
fraud = dataset[dataset.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

UnderSampling:
Buliding a sample dataset containing comparable legit and fraud transaction

```python
legit_sample = legit.sample(n=492)
```

```python
legit_sample.shape
```

Concatinate the two dataframes

```python
new_dataset = pd.concat([legit_sample,fraud], axis=0)
```
