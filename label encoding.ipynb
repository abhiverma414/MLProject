Label Encoding:

--Converting the labels into numeric form

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

Label Encoding of breast cancer dataset

```python
cancer_data = pd.read_csv('data.csv')
```

```python
cancer_data.head()
```

```python
cancer_data['diagnosis'].value_counts()
```

```python
#Load the lable encoder function
label_encoder = LabelEncoder()
```

```python
labels = label_encoder.fit_transform(cancer_data.diagnosis)
```

```python
#appending the labels to DataFrame
cancer_data['targets'] = labels
```

```python
cancer_data.head()
```
